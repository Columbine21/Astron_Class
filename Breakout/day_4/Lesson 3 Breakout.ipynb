## 1) Plotting Residuals

In this assignment we will make a basic plot showing a residual. Below is code that generates reproducably the same noisey data

```python
import numpy as np
np.random.seed(10) # This will make us always get the same random numbers

x = np.linspace(0,1,30)
data = np.random.rand(30) + x**2

%pylab inline
plt.plot(x,data)
```

1) Overplot the true solution ($y=x^2$)

```python
# construct the data 
np.random.seed(10) # This will make us always get the same random numbers

x = np.linspace(0,1,30)
y_noisy = np.random.rand(30) + x**2
y_base = x ** 2
y_diff = y_noisy - y_base
```

```python
%pylab inline
plt.plot(x, y_noisy)
plt.plot(x, y_base)
plt.show()
```

2) Plot the difference between the true solution (i.e. data - model)

```python
%pylab inline
# set title
# plt.title(r'$\mathrm{Histogram\ of\ Gaussian:}\ \mu=100,\ \sigma=15$')
plt.title(r'differece', color = 'black')

plt.plot(x, y_noisy, color = 'blue', label = 'real', linewidth =0.5)
plt.plot(x, y_base, color = 'green', label = 'baseline', linewidth =0.5)
plt.plot(x, y_diff, color = 'red', label = "difference", linewidth =0.5)
plt.legend(loc='upper right', fontsize = 12)

plt.ylim(0, 1.6)
# plt.xlim(0, 1.1)
plt.grid()
```

```python
array_list = np.array([y_base, y_noisy, y_diff])
array = np.vstack([y_base, y_noisy, y_diff])
d = {0 : "y_base", 1 : "y_noisy", 2 : "diff"}
c = {0 : "red", 1 : "green", 2 : "blue"}
```

3) Get both of these onto the same image using a subplot command.

```python
fig, axes = plt.subplots(3, 1, figsize=(6,8))
for n,i in enumerate(axes):
    i.plot(x, array[n], label = d[n], color = c[n], linewidth = 1)
    i.legend()
plt.show()
```

## 2) Pillars of Creation

In this folder is another PNG file of a classic Hubble image, Pillars of Creation.

1) Open this file using the "imread" command.

2) Plot the image in each color.

3) Make a histogram showing all three colors intensity. Plot them on the same plot.

```python
import matplotlib.image as mpimg

import matplotlib.pylab as plt

img = mpimg.imread("./pillars_of_creation.png")

colorList = {0 : "red", 1 : "blue", 2 : "green"} 
```

```python
fig, axes = plt.subplots(3, 1, figsize=(6,8))
# plt.subplots_adjust(hspace = 0.3)
plt.subplots_adjust(bottom = 0)
for n,i in enumerate(axes):
    i.imshow(img[:,:,n])
    i.set_title(colorList[n])
```
