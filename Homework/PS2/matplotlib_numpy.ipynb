## Problem 3) Intro to Numerical Integration (10 Points)

To preview next week's lesson, lets play around with some basic numerical integration techniques. Consider the following function;

$$E(x) = \frac{2}{\sqrt{\pi}}\int_{0}^{x}e^{-t^{2}}dt.$$

The integral within this function can not be done analytically; i.e. it doesn't possess a nice functional solution like integrals encountered in the standard calculus course. Instead we will have to use numerical integration techniques to solve for it.

1) Code up a function for $E(x)$, you may use the basic trapizoidal integration solver below. Plot the function for values between $x=1$ and $x=100$.

```python
def func_E(x):
    return 2 * (np.e ** (- x ** 2)) / sqrt(np.pi)

def trap_int(func, a, b, N=1000):
    """
    very simple trapezoidal integration solver
    """
    # bin width
    h = (b-a)/float(N)
    
    # get x-values
    x_range = np.linspace(a, b, N+1)

    # get y-values
    y_range = func(x_range)

    # evaluate trapezoidal sum
    I = h*(0.5*y_range[0] + 0.5*y_range[-1] + sum(y_range[1:-1]))
    
    return I

fig = plt.figure(figsize=(6,6))
X = np.linspace(1, 100, 100)
trap_int(func_E, 0, 3, 10000)
Y = [trap_int(func_E, 0, x, 10000) for x in X]
plt.plot(X, Y, color = 'blue', label = r'$E(x) = \frac{2}{\sqrt{\pi}}\int_{0}^{x}e^{-t^{2}}dt.$')
plt.xlim(0,20)
plt.xlabel('x')
plt.ylabel('E(x)')
plt.legend()
```

$$E(1.5) =  0.96610514(6)$$

```python
# initialize the varible
n = 1000

# search the min n satisfies the accuracy 
while(True):
    if(abs(trap_int(func_E, 0, 0.5, n) - 0.5204998775) <= 0.0000000005 and
       abs(trap_int(func_E, 0, 1, n) - 0.8427007925) <= 0.0000000005 and 
       abs(trap_int(func_E, 0, 1.5, n) - 0.9661051465) <= 0.0000000005):
        break
    else:
        n = n + 1
        
# output the result
print("The minimum n is {}".format(n))
print("with n = {}, E(0.5) = {}".format(n, trap_int(func_E, 0, 0.5, n)))
print("with n = {}, E(1.0) = {}".format(n, trap_int(func_E, 0, 1, n)))
print("with n = {}, E(1.5) = {}".format(n, trap_int(func_E, 0, 1.5, n)))
# print(trap_int(func_E, 0, 0.5, n), trap_int(func_E, 0, 1, n), trap_int(func_E, 0, 1.5, n))
```

Assume a box size with units between 0 and 1. Have velocities be in each direction have a Gaussian distribution with mean 0 and standard deviation 0.01. Plot the field!

```python
import matplotlib.pylab as plt
import numpy as np

fig = plt.figure(figsize=(6,6))
global particle_pos
particle_pos = np.random.random([1000, 2])
particle_velocities = np.random.normal(0.0, 0.01, [1000, 2])

# to make our field more beautiful we use random color & size to draw the point.
color_seed = (20 * np.random.rand(1000))**2  # 0 to 10 point radii
size_seed = (3 * np.random.rand(1000))**2


plt.scatter(particle_pos[:, 0], particle_pos[:, 1],marker='^',s = size_seed, c=color_seed)
plt.figure()
```

2) Write a function to calculate the force on each particle. Remember that force is a vector quanity.

```python
from scipy import constants as C
# Assume that all particles are unit mass

# construct the dist matrix of each pair of particles
dist_matrix = np.array([
    [np.sqrt((x-x1)**2 + (y-y1)**2) for n,(x1, y1) in enumerate(particle_pos)] 
    for n,(x,y) in enumerate(particle_pos)])
```

```python
def calculate_force():
    #initialize the array
    force_x = []
    force_y = []
    # calculate the Magnitude of the force
    for i in range(1000):
        xi, yi = 0, 0
        for j in range(1000):
            if(i == j): continue
            else:
                xi += C.G * (particle_pos[i,0] - particle_pos[j,0]) / (dist_matrix[i,j]**3) 
                yi += C.G * (particle_pos[i,1] - particle_pos[j,1]) / (dist_matrix[i,j]**3) 
        force_x.append(xi)
        force_y.append(yi)
    force_x = np.array(force_x).reshape(1000,-1)
    force_y = np.array(force_y).reshape(1000,-1)
    force = np.hstack((force_x, force_y))
    return force
```

```python
force = calculate_force()
```
